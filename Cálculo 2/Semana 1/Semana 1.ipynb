```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir os valores de x e y
x = np.linspace(-10, 10, 100)  # Intervalo de -10 a 10
y = np.linspace(-10, 10, 100)

# Criar uma grade de coordenadas
X, Y = np.meshgrid(x, y)

# Calcular Z = x^2 + y^2
Z = X**2 + Y**2

# Criar a figura e o eixo 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plotar a superfície
surface = ax.plot_surface(X, Y, Z)

# Adicionar uma barra de cores
fig.colorbar(surface, ax=ax)

# Títulos e rótulos
ax.set_title('Gráfico de Superfície 3D: Z = x^2 + y^2')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_parabola(a, b, c):
    """
    Plota uma parábola baseada nos coeficientes a, b e c.
    """
    x = np.linspace(-10, 10, 500)  # Gera 500 valores de x entre -10 e 10
    y = a * x**2 + b * x + c       # Calcula os valores de y
    
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f'y = {a}x² + {b}x + {c}')
    plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Eixo x
    plt.axvline(0, color='black', linewidth=0.8, linestyle='--')  # Eixo y
    plt.title('Gráfico da Parábola')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid()
    plt.show()

# Exemplo de uso
plot_parabola(1, -3, 2)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import math

# Criando os valores de x e y
x = np.linspace(0, 10, 10)  # Gera 10 pontos entre 0 e 10
y = np.linspace(0, 10, 10)

# Criando um grid 2D
X, Y = np.meshgrid(x, y)

# Calculando Z para cada par (x, y)
Z = np.sin(X + Y)  # Função de exemplo

# Criando o DataFrame (opcional)
df = pd.DataFrame({'x': X.ravel(), 'y': Y.ravel(), 'z': Z.ravel()})

# Criando a figura
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Criando a superfície
surface = ax.plot_surface(X, Y, Z, cmap='viridis')

# Adicionando a barra de cores
fig.colorbar(surface, ax=ax)

# Exibir o gráfico
plt.show()
```
